```python
plt.figure(figsize=(15,300))
sns.histplot(data=df, x="purchase_lead", binwidth=20,kde=True)
```

```python
(df.purchase_lead >300).value_counts()
```

There are a few bookings that were done more than 2 years before the travel date and you suspect that they might be outliers, we  can consider removing them from your dataset before building your predictive model. However, before doing that, it's important to investigate the reason behind these bookings.

As it shows, it's possible that these bookings were made far in advance due to cancellations and rebookings within a period of 6 months. In this case, these bookings may not necessarily be outliers and removing them from the dataset may not be the best approach. Instead, we can consider adding a new feature to your dataset that indicates if a booking was made due to a cancellation and rebooking.

On the other hand, if you find that these bookings were made in error or due to some other reason that is not representative of your normal booking patterns, then you can consider removing them from the dataset. However, it's important to ensure that you are not introducing any bias into your model by removing these outliers.

If we assume that no customer is booking in advance of more than 1 and half year we will remove all entries with purchase_lead more than 300 days.

# One Hot Encoding

```python
countries = pd.read_csv("Countries.csv",  encoding="ISO-8859-1")
countries
```

```python
df_final = df.merge(countries, left_on='booking_origin', right_on='Country',)
df_final
```

```python
df_final.Continent.unique()
```

```python
%pip install scikit-learn
from sklearn.preprocessing import OneHotEncoder


#create instance of one hot encoder
encoder = OneHotEncoder(handle_unknown='ignore')

#one hot encode trip type
encoder_df = pd.DataFrame(encoder.fit_transform(df_final[["trip_type"]]).toarray())
encoder_df = encoder_df.rename(columns={0:'RoundTRip', 1:'OneWayTrip',2:'CircleTrip'})
df_final = df_final.join(encoder_df)
```

```python
encoder_df = pd.DataFrame(encoder.fit_transform(df_final[["Continent"]]).toarray())
encoder_df = encoder_df.rename(columns={0 : 'Oceania' , 1 : 'Asia', 2 : 'Europe', 3 : 'South America', 4 : 'Africa' ,5 :'North America'})
df_final = df_final.join(encoder_df)
```

```python
#drop categorical columns now
df2 = df_final.drop(['trip_type','booking_origin','Country','route','Continent', "flight_day"], axis=1)
df2
```

```python
label = df2['booking_complete']
df2 = df2.drop('booking_complete', axis=1)
df2
```

```python
label
```

# Normalizaing the values

```python
from sklearn.preprocessing import StandardScaler

#create a standard scaler object
scaler = StandardScaler()

#fit and transform the data
scaled_df = scaler.fit_transform(df2)
```

```python
scaled_df = pd.DataFrame(scaled_df, columns = df2.columns)
```

```python
scaled_df['label']  = label
```

```python
scaled_df
```

```python
corr = scaled_df.corr()

plt.figure(figsize=(10,7))

#plot the heatmap
sns.heatmap(corr)
```

```python
from sklearn.model_selection import train_test_split

X = scaled_df.iloc[:,:-1]
y = scaled_df['label']

X_train, X_test, y_train, y_test = train_test_split(X.to_numpy(), y.to_numpy(), test_size=0.20, random_state=42)
```

```python
print('X_train shape: ', X_train.shape)
print('X_test shape: ', X_test.shape)
print('y_train shape: ', y_train.shape)
print('y_test shape: ', y_test.shape)
```

```python
%pip install --upgrade scikit-learn
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.neural_network import MLPClassifier
from sklearn import svm
from sklearn.svm import SVC
%matplotlib inline
from sklearn.metrics import classification_report, f1_score, precision_score, confusion_matrix
from sklearn.metrics import plot_confusion_matrix
from sklearn.model_selection import train_test_split,GridSearchCV,cross_val_score


import warnings
warnings.filterwarnings(action="ignore")
```

```python
models = {
    "     K-Nearest Neighbour" : KNeighborsClassifier(),
    "Decision Tree CLassfiier" : DecisionTreeClassifier(),
    "           Random Forest" : RandomForestClassifier(),
    "                     ANN" : MLPClassifier()
 }
```

```python
print("Training accuracy :")
for name, model in models.items():
  model.fit(X_train,y_train)
  print(name + ": {:.2f}%".format(model.score(X_train,y_train) * 100))
  print("Test Accuracy")
  y_pred = model.predict(X_test)
  print("Accuracy", model.score(X_test,y_test) )
  print('Precision:', precision_score(y_test, y_pred))
  print('F1-Score:', f1_score(y_test, y_pred))
  print('\n')

```

```python
main_model_rfc = RandomForestClassifier()

params={
            'max_depth'               : [4,6,8,12], 
            'max_features'            : [6,7,8], 
            'min_samples_leaf'        : [2, 4], 
            'min_samples_split'       : [2, 5, ], 
            'n_estimators'            : [ 75,100,200],
            'criterion'               : [ "gini", "entropy" ]
}

random_search_rfc = GridSearchCV(
                                       estimator=RandomForestClassifier(class_weight='balanced'), 
                                       param_grid=params, 
                                       cv=5, 
                                       verbose=3,
                                       return_train_score= True
                                       )
random_search_rfc.fit(X_train,y_train)
```

```python
random_search_rfc.best_estimator_
```

```python
random_search_rfc.best_params_
```

```python
model_rfc = RandomForestClassifier(
                                   class_weight      = 'balanced',
                                   bootstrap         = False,
                                   criterion         = 'entropy',
                                   max_depth         = 12, 
                                   max_features      = 8,
                                   min_samples_leaf  = 2, 
                                   min_samples_split = 2,
                                   n_estimators      = 200)
m = model_rfc.fit(X_train,y_train)
score_rfc = cross_val_score(model_rfc,X_train,y_train,cv=10)
print("Training:\nAverage score for 10CV is : ", score_rfc.mean())

y_pred_train_rfc = model_rfc.predict(X_test)
print("\nAccuracy score of test data is: ", model_rfc.score(X_test,y_test) )
print(classification_report(y_test, y_pred_train_rfc ))
```

```python
print(confusion_matrix(y_test,y_pred_train_rfc))
```

```python
plt.figure(figsize=(10,5))
sorted_idx = model_rfc.feature_importances_.argsort()
plt.barh(scaled_df.iloc[:,:-1].columns[sorted_idx], model_rfc.feature_importances_[sorted_idx])
plt.xlabel("Random Forest Feature Importance")
```

# Balancing the dataset


```python
scaled_df
```

```python
scaled_df.label.value_counts()
```

```python
scaled_df_0 = scaled_df[scaled_df.label ==0].sample(n=7000)
scaled_df_0
```

```python
scaled_df_new = pd.concat([scaled_df[scaled_df.label==1], scaled_df_0], ignore_index=True)
scaled_df_new
```

```python
scaled_df_new = scaled_df_new.sample(frac = 1).reset_index(drop=True)
scaled_df_new
```

```python
X = scaled_df_new.iloc[:,:-1]
y = scaled_df_new['label']

X_train, X_test, y_train, y_test = train_test_split(X.to_numpy(), y.to_numpy(), test_size=0.20, random_state=42)
```

```python
main_model_rfc = RandomForestClassifier()

params={
            'max_depth'               : [10,12,15], 
            'min_samples_leaf'        : [2, 4], 
            'min_samples_split'       : [2, 3], 
            'n_estimators'            : [ 75,100,200],
}

random_search_rfc = GridSearchCV(
                                       estimator=RandomForestClassifier(class_weight='balanced'), 
                                       param_grid=params, 
                                       cv=5, 
                                       verbose=3,
                                       return_train_score= True,n_jobs=-1
                                       )
random_search_rfc.fit(X_train,y_train)
```

```python
random_search_rfc.best_params_
```

```python
random_search_rfc.best_estimator_
```

```python
model_rfc = RandomForestClassifier(class_weight='balanced', max_depth=12,
                       min_samples_leaf=4, min_samples_split=3,
                       n_estimators=75)
m = model_rfc.fit(X_train,y_train)
score_rfc = cross_val_score(model_rfc,X_train,y_train,cv=10)
print("Training:\nAverage score for 10CV is : ", score_rfc.mean())

y_pred_train_rfc = model_rfc.predict(X_test)
print("\nAccuracy score of test data is: ", model_rfc.score(X_test,y_test) )
print(classification_report(y_test, y_pred_train_rfc ))
```

```python
# y_true is the true labels, y_pred is the predicted labels
cm = confusion_matrix(y_test, y_pred_train_rfc)

# create a heatmap visualization of the confusion matrix
sns.heatmap(cm, annot=True,fmt='d', cmap="Blues")

# set the axis labels and title
plt.xlabel("Predicted Labels")
plt.ylabel("True Labels")
plt.title("Confusion Matrix")

# display the plot
plt.show()
```
